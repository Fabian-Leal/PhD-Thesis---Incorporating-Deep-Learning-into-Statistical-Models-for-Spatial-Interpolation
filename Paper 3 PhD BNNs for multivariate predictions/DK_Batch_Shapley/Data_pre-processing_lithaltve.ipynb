```python
density_data = pd.read_csv("../Curated_data/cluster_0_data.csv", low_memory=False)
```

```python
# In the other notebook

# Load the indexes from the CSV file
indexes = pd.read_csv('indexes.csv', squeeze=True)

# Assuming you have another DataFrame called another_df


# Select rows from another_df using the indexes
selected_rows = density_data.loc[indexes]

# Display the selected rows
selected_rows
```

```python
# Scale X,Y,Z using the same scale 
columns_to_normalize = ['X', 'Y', 'Z']   

min_values = selected_rows[columns_to_normalize].min()

for column in columns_to_normalize:
    selected_rows[column] = selected_rows[column] - min_values[column]

# Calculate the minimum and maximum values from all three variables
min_values = selected_rows[columns_to_normalize].min().min()
max_values = selected_rows[columns_to_normalize].max().max()

# Scale the data to (0, 1) using the calculated minimum and maximum values
for column in columns_to_normalize:
    selected_rows[column] = (selected_rows[column] - min_values) / (max_values - min_values)

#Scale the rest of the covariates in their own scale

```

```python
for col in selected_rows.columns:
    print(col)
```

```python
alteration_keywords = ["Alt"]
lithology_keywords = ["Lith", "Orig_Lith"]
vein_keywords = ["Vein"]

# Function to check if a column name contains any of the keywords
def contains_keyword(col, keywords):
    return any(keyword in col for keyword in keywords)

# Select columns based on keywords
alteration_columns = [col for col in selected_rows.columns if contains_keyword(col, alteration_keywords)]
lithology_columns = [col for col in selected_rows.columns if contains_keyword(col, lithology_keywords)]
vein_columns = [col for col in selected_rows.columns if contains_keyword(col, vein_keywords)]

# Combine all selected columns
selected_columns = alteration_columns + lithology_columns + vein_columns + ['X','Y','Z']

# Display the selected columns
print(selected_columns)
```

```python
dflitaltve = selected_rows[selected_columns]
```

```python
for col in dflitaltve.columns:
    print(col, dflitaltve[col].notna().sum())
```

```python
import pandas as pd
import plotly.graph_objs as go
from plotly.offline import iplot, init_notebook_mode


# Find the top 7 Lith1_Code based on their frequency
top_7_lith_codes = dflitaltve['Lith1_Code'].value_counts().nlargest(7).index

# Filter the DataFrame to only include these top 7 Lith1_Code
filtered_df = dflitaltve[dflitaltve['Lith1_Code'].isin(top_7_lith_codes)]

# Define colors for each Lith1_Code
color_map = {
    'UWBO': 'cyan', 'UPXO': 'blue', 'MPE': 'green', 'MPHB': 'yellow', 'UPX': 'red', 'IGB': 'magenta', 'UWBOP': 'orange'
}

# Create Plotly figure
fig = go.Figure()

# Add scatter plot
scatter = go.Scatter3d(
    x=filtered_df['X'],
    y=filtered_df['Y'],
    z=filtered_df['Z'],
    mode='markers',
    marker=dict(
        size=4,
        color=[color_map[lith] for lith in filtered_df['Lith1_Code']],  # Color based on Lith1_Code
        opacity=0.8
    )
)
fig.add_trace(scatter)

# Set labels and title
fig.update_layout(
    title='Top 7 Lith1_Code in Space',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    ),
    legend_title="Lith1_Code"
)

# Create a legend (manually, since Plotly doesn't automatically create legends for scatter3d)
for lith, color in color_map.items():
    if lith in top_7_lith_codes:
        fig.add_trace(go.Scatter3d(
            x=[None], y=[None], z=[None],
            mode='markers',
            marker=dict(
                size=5,
                color=color,
                opacity=0.8
            ),
            showlegend=True,
            name=lith
        ))

init_notebook_mode(connected=True)
iplot(fig)
```
